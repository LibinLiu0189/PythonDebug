### 每年最受欢迎的电影类别是哪些？

```python
# 获取genres为null的行的索引
df_null_index = df[df['genres'].isnull()].index
df_null_index
```

```python
# 删除genres为null的行
df.drop(df_null_index, inplace=True)
```

```python
# 15年最受欢迎的电影分布
df.query('release_year == 2015')['popularity'].describe()
```

```python
# 查询2015年且受欢迎度大于75%的电影
df_15 = df.query('release_year == 2015 and popularity > 0.964706')

df_15
```

```python
# 统计2015年且受欢迎度大于75%的电影
genres = []
# 对genres列，以'\'分组得到list，取list中的每一个元素添加到genres
df_15['genres'].apply(lambda x: [genres.append(y) for y in x.split('|')])
# 统计各个类型出现的频率
genres_frq = pd.Series(genres).value_counts()
genres_frq.plot(kind="bar", title="the most popular movie genre of 2015",color='green', legend=True, label="genres freq");
genres_frq
```

## 探索性数据分析

> **提示**：整理和清理数据之后，现在可以进行探索。计算统计值，创建视图，解决你在简介段提出的研究问题。建议采用系统化方法。一次探索一个变量，然后探索变量之间的关系。

### 1. vote_average和revenue是否呈正相关

```python
# 查看revenue分布
df['revenue_adj'].describe()
```

```python
# 收入为0的电影的评分
df.query('revenue_adj == 0')['vote_average'].describe()
```

```python
# 收入大于0的电影的评分
df.query('revenue_adj > 0')['vote_average'].describe()
```

```python
# 获取收入大于0时的收入，以及其对应的评分
y_revenue = df.query('revenue > 0')['revenue']
#y_revenue
x_vote_average = df.query('revenue > 0')['vote_average']
#x_vote_average
```

```python
#该df指定x,y变量绘制散列图
df.query('revenue_adj > 0').plot(x='vote_average',y='revenue_adj',kind='scatter');
```

1. 随着平均投票逐渐升高，票房收入也逐渐增加，并且高票房收入的电影都处于平均投票高的区域。

### 2. popularity和revenue是否呈正相关

```python
#该df指定popularity,revenue 大于0 时绘制散列图
df.query('popularity > 0 and revenue_adj > 0').plot(x='popularity',y='revenue_adj',kind='scatter');
```

2. 此时并未类似vote_average和revenue呈明显正相关的分布，票房高的电影也有受欢迎度低，相反，票房不高的电影也有受欢迎度高的

### 3. budget和revenue是否呈正相关

```python
df['budget_adj'].describe()
```

```python
# 将预算为0的行删除
# df.drop(df.query('budget == 0').index,inplace=True)
# 再次确认数据被删除
# df.query('budget == 0').index
```

```python
df.query('budget_adj > 0 and revenue_adj > 0').plot(x='budget_adj',y='revenue_adj',kind='scatter');
```

3. 此时的情况类似popularity和revenue，图中预算低的情况时出现了票房高的情况，从绝大多数来看，随着预算的增加票房分布呈上升的趋势，在预算1.5后票房分布逐渐上升。

## 结论
1. 平均投票与票房：随着平均投票逐渐升高，票房收入也逐渐增加，并且高票房收入的电影都处于平均投票高的区域。

3. 预算与票房：此时的情况类似popularity和revenue，图中预算低的情况时出现了票房高的情况，从绝大多数来看，随着预算的增加票房分布呈上升的趋势，在预算1.5后票房分布逐渐上升。
